```python
import pandas as pd
import glob
import os
import matplotlib.pyplot as plt
```

```python
path = "/Users/wiesner/Dev/PureEdgeSim/PureEdgeSim/output/2019-12-19_16-21-58"
dfs = [pd.read_csv(csv) for csv in glob.glob(os.path.join(path, "*.csv"))]
full_df = pd.concat(dfs)
full_df = full_df.sort_values(["Orchestration architecture", "Edge devices count"])
full_df = full_df.reset_index(drop=True)
full_df = full_df.drop(["Dead devices count", "Average remaining power (%)", "First edge device death time (s)", "List of remaining power (%) (only battery powered devices / 0 = dead)", "List of the time when each device died (s)"], axis=1) 
full_df
```

```python
y_max = full_df["Energy consumption (Wh)"].max()
```

```python
def plot_me(ar):
    fig, ax = plt.subplots()
    all_df = full_df[full_df["Orchestration architecture"] == ar]
    all_df.plot(kind='line', x='Edge devices count', y='Energy consumption (Wh)', ax=ax)
    all_df.plot(kind='line', x='Edge devices count', y='Cloud energy consumption (Wh)', ax=ax)
    all_df.plot(kind='line', x='Edge devices count', y='Fog energy consumption (Wh)', ax=ax)
    all_df.plot(kind='line', x='Edge devices count', y='Edge energy consumption (Wh)', ax=ax)
    plt.xlim([100, 500])
    plt.ylim([0, y_max])
    plt.show()
```

```python
plot_me("ALL")
```

```python
plot_me("CLOUD_ONLY")
```
